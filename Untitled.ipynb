```python
import pandas as pd


df = pd.read_csv('data.csv')

df.info()
```

```python
#произвожу работу с пропущеными значениями

df['Purchaser Codes'] = df['Purchaser Codes'].fillna('unknown') #заполняю текстовыми значениями пропуски столбца
df = df.dropna()#удаляю путсые значения

df


```

```python
#очистка от аномалий

int_cols = df.select_dtypes(include='int64').columns#выбераю числовые данные

#очистка от аномалий методом квантилей
for column in int_cols:
    lower_bound = df[column].quantile(q=0.025)
    upper_bound = df[column].quantile(q=0.975)
    mask = (df[column] >= lower_bound) & (df[column] <= upper_bound)

    clean = df[mask].dropna()

clean


```

```python
clean.info()
```

```python
#преобразовал тип данных
clean ['Production Date Entered'] = clean['Production Date Entered'].astype('datetime64')

clean.info()
```
